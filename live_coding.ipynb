```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

numerical_features = df.select_dtypes(['number']).columns
preprocessor = Pipeline([
    ('scaler', StandardScaler())
])

X = df.drop('Etat de charge SOC(%)', axis=1)
y = df['Etat de charge SOC(%)']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.transform(X_test)


```

```python
import tensorflow as tf

model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_squared_error'])

history = model.fit(X_train, y_train, epochs=50, validation_split=0.2, batch_size=32)
```

```python
loss, mse = model.evaluate(X_test, y_test)
print(f"Mean Squared Error: {mse}")
```

```python
plt.plot(history .history['mean_squared_error'], label='mean_squared_error')
plt.plot(history.history['val_mean_squared_error'], label = 'val_mean_squared_error')
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label = 'val_loss')

plt.xlabel('Epoch')
plt.legend()
plt.show()
```

```python
lstm_model = tf.keras.models.Sequential([
    tf.keras.layers.LSTM(64, activation='relu', return_sequences=True, input_shape=(X_train.shape[1], 1)),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.LSTM(64, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1)
])


lstm_model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_squared_error'])

history = lstm_model.fit(X_train, y_train, epochs=50, validation_split=0.2, batch_size=32)
```

```python
loss, mse = model.evaluate(X_test, y_test)
print(f"First model Mean Squared Error: {mse}")

loss, mse = lstm_model.evaluate(X_test, y_test)
print(f"LSTM model Mean Squared Error: {mse}")
```

```python
plt.plot(history .history['mean_squared_error'], label='mean_squared_error')
plt.plot(history.history['val_mean_squared_error'], label = 'val_mean_squared_error')


plt.xlabel('Epoch')
plt.legend()
plt.show()
```

```python
lstm_model_avec_regularization = tf.keras.models.Sequential([
    tf.keras.layers.LSTM(64, activation='relu', return_sequences=True, input_shape=(X_train.shape[1], 1), kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.LSTM(64, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(32, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1, kernel_regularizer=tf.keras.regularizers.l2(0.01))
])

lstm_model_avec_regularization.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_squared_error'])

history = lstm_model.fit(X_train, y_train, epochs=50, validation_split=0.2, batch_size=32)
```

```python
loss, mse = model.evaluate(X_test, y_test)

print(f"First model Mean Squared Error: {mse}")
loss, mse = lstm_model.evaluate(X_test, y_test)
print(f"LSTM model Mean Squared Error: {mse}")

loss, mse = lstm_model_avec_regularization.evaluate(X_test, y_test)
print(f"LSTM model with regularization Mean Squared Error: {mse}")
```
